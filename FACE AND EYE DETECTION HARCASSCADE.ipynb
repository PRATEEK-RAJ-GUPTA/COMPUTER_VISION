FUNCTION TOADISPLAY IMAGE

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def imshow(title="Image",image=None,size=10):
    w,h=image.shape[0],image.shape[1]
    aspect_ratio=w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')

image=cv2.imread('Images/women_pics.jpg')
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

imshow('original_photo',image)

face=face_classifier.detectMultiScale(gray,scaleFactor=1.3,minNeighbors=5)
#returns a tuple of argument
if face is ():
    print('No face detected')
    

for (x,y,w,h) in face:
    cv2.rectangle(image,(x,y),(x+w,y+h),(1,0,255),5)

imshow('FACE DETECTED',image)
```

EYE DETECTION 

```python
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')

image = cv2.imread('Images/women_pics.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

face=face_classifier.detectMultiScale(gray,scaleFactor=1.3,minNeighbors=5)

if face is ():
    print('No face detected')

    
for (x,y,w,h) in face:
    cv2.rectangle(image,(x,y),(x+w,y+h),(1,0,255),5)
    para_gray=gray[y:y+h,x:x+w]
    para_color=image[y:y+h,x:x+w]
    eyes = eye_classifier.detectMultiScale(para_gray, 1.3, 8)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(para_color,(ex,ey),(ex+ew,ey+eh),(255,255,0),5)
        
imshow('EYE AND FACE DETECTED',image)

```

EYE DETECTION IN REAL TIME TAKEN IMAGE(IMAGE TAKEN THORUGH WEBCAM)

```python
import cv2
import numpy as np
face_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('Haarcascades/haarcascade_eye.xml')

def face_detector(image,size=0.5):
    gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray, 1.1, 5)
    if faces is ():
        return image
    
    for (x,y,w,h) in faces:
        cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),2)
        para_gray = gray[y:y+h, x:x+w]
        para_color = image[y:y+h, x:x+w]
        eyes = eye_classifier.detectMultiScale(para_gray)
        
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(para_color,(ex,ey),(ex+ew,ey+eh),(0,0,255),1)
            
    
    return para_color

# command to open wecam

cap=cv2.VideoCapture(0)

while True:#while webcam is open
    
    ret,frame=cap.read()
    cv2.imshow('OUR FACE EXTRACTOR', face_detector(frame))
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
        # press q to close and 1 to pause
    
cap.release()
cv2.destroyAllWindows()
    
    
```
